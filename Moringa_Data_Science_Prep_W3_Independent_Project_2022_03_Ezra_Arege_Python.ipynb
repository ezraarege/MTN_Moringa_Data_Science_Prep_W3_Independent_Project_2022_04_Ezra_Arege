#### 2. DATA CLEANING

1.Duplicates


2.Missing data


Get rid of redudant columns in all our dataset

```python
desc.columns
```

```python
# DROP REDUDANT COLUMNS

desc =  desc.drop(['Unnamed: 0', "STATUS", "LATITUDE", "LONGITUDE", "AREA", "ZONENAME", "DECOUPZONE"], axis = 1)
```

```python
desc.rename(columns = {"VILLES" : "CITY"}, inplace=True)
```

```python
#drop null values

desc.dropna()
```

Lets check whether there are duplicates

```python
# sum of duplicates in dataset

desc.duplicated().sum()
```

```python
#delete duplicated values

desc.drop_duplicates(subset = ["SITE_CODE", 'CELL_ID'], inplace = True)

desc.head()
```

```python
desc.shape
```

```python

df1 = df1.drop(["CELL_ON_SITE", 'DW_A_NUMBER_INT', 'DW_B_NUMBER_INT', 'COUNTRY_A', 'COUNTRY_B'], axis = 1)
```

```python
df1.columns
```

```python
df2.columns
```

```python
 # delete redundant columns 

 
df2 = df2.drop(["CELL_ON_SITE", 'DW_A_NUMBER', 'DW_B_NUMBER', 'COUNTRY_A', 'COUNTRY_B'], axis = 1) 
```

```python
df2.columns
```

```python
df3.columns
```

```python
# delete redundant columns 

df3 = df3.drop(["CELL_ON_SITE", 'DW_A_NUMBER_INT', 'DW_B_NUMBER_INT', 'COUNTRY_A', 'COUNTRY_B'], axis = 1)
```

```python
df1.columns
```

```python
# rename misspelled columns

df1 = df1.rename(columns = {"PRODUTC": "PRODUCT",  "DATETIME" : "DATE_TIME"})
df3 = df3.rename(columns = {"SIET_ID": "SITE_ID", "CELLID" : "CELL_ID"})
```

```python
df1.duplicated().sum()
df2.duplicated().sum()
df3.duplicated().sum()
```

```python
df3.columns
```

```python
# delete duplicates
df1.drop_duplicates(subset =["CELL_ID", "SITE_ID"], inplace = True)
df2.drop_duplicates(subset =["CELL_ID", "SITE_ID"], inplace = True)
df3.drop_duplicates(subset =["CELL_ID", "SITE_ID"], inplace = True)


#print no of remaining columns

print(df1.shape)
print(df2.shape)
print(df3.shape)
```

##### Merge our datasets ready for analysis

```python
# concatenate our CDR datasets chronologically

df = pd.concat([df1, df2, df3])
```

```python
df.columns
```

```python
desc.rename(columns = {"SITE_CODE" : "SITE_ID"}, inplace = True)
```

```python
data = pd.merge(df, desc, on = "SITE_ID", how = "inner")
data.head()
```

```python
data.tail()
```

```python
data.isnull().sum()
```

## Data Exploration

```python
### Which ones were the most used city for the three days


data["CITY"].value_counts().head(20)
```

```python
#Which cities were the most used during business and home hours?

data.groupby(['CITY', "DATE_TIME"])['VALUE'].sum().sort_values(ascending=False)
```

```python
data["CITY"].value_counts().head(20).plot(kind='bar', figsize=(10,10))
```

```python
# Which regions have the highest frequency of calls sms data
data['REGION'].value_counts()
```

```python
data['REGION'].value_counts().plot(kind='bar', figsize=(10,10))
```

```python

product_use = plt.figure(figsize=(5,4))
use = data["PRODUCT"].value_counts()
sns.barplot(x= use, y=use.index)
plt.xlabel('Product Usage')
```

```python
# which city brings most revenue

df4 = data.groupby(['CITY'])['VALUE'].sum().sort_values(ascending=False)
```

```python
df4.head(20)
```

```python
#Which city brings in more money?
data.groupby(['CITY'])['VALUE'].sum().sort_values(ascending=False).head(15).plot(kind="bar", figsize =(18,10))
```
